<hr style="border:0.02in solid gray"> </hr>

> <font color= 'darkcyan' > <font size = 3> En el presente trabajo, se pretende realizar dos estrategias de inversión una pasiva y una activa, esto con la finalidad de observar cuál será la mejor opción para el millón de pesos que pone en juego el inversionista en curso. Todo esto será posible mediante el uso de un ETF.
>
> <font color= 'darkcyan' > <font size = 3> Para iniciar cabe destacar que un índice bursátil es un valor numérico, que se calcula según los precios de mercado de cada uno de los valores que componen ese índice en un momento determinado. La rentabilidad de un índice es la variación de su valor de un periodo a otro. 
    En pocas palabras, reflejan el comportamiento del mercado accionario o de diferentes empresas con las que se comparten similitudes operativas.
>   
> <font color= 'darkcyan' > <font size = 3> Se utilizan como:
>
> <font color= 'darkcyan' > <font size = 3> - Indicadores de desempeño de una economía.
>
> <font color= 'darkcyan' > <font size = 3> - Medidas de aversión al riesgo.
>
> <font color= 'darkcyan' > <font size = 3> - Como referencia para estrategias de inversión.
>
> Posteriormente tenemos que:
>
> La inversión pasiva consiste en invertir en bolsa utilizando ETFs o fondos indexados, fondos que replican el comportamiento de cualquier índice bursátil (por ejemplo, podrías adquirir un fondo que replique el comportamiento del IBEX 35). Para este caso, se está haciendo uso del ETF IShare NAFTRAC mismo que busca replicar al índice del S&P500/BMV.
>
> La inversión activa se dedica a elegir acciones de forma individual para intentar batir la rentabilidad media del mercado.

> <font color= 'darkcyan' > <font size = 3> Cabe desatacar que se cuenta con un capital de \$1,000,000 millón de pesos el cual se utilizará en la inversión procedente. Asimismo el porcentaje de comisiones que se cobrara con cada transacción será de 0.125\% sobre las ventas.

```python
# Capital
k = 1000000
# Comision
c = 0.00125
```

```python
# Path
abspath = path.abspath('files')
```

```python
# Files
files = fn.f_files(abspath)
```

```python
# Get into files
data_files = fn.f_datafiles(files)
```

```python
# Extract dates
dates = fn.f_dates(files)
```

```python
# Names in the BMV
tickers = fn.f_tickers(files, data_files)
```

# <font color= #6B6B6B> 5. Passive investment </font>

<hr style="border:0.02in solid gray"> </hr>

> <font color= 'darkcyan' > <font size = 3> Se realiza la descarga de datos desde yahoo finance y se obtiene lo siguiente.

```python
# Download data
data_close = fn.f_down_data(tickers)
data_close.head(5)
```

> <font color= 'darkcyan' > <font size = 3> Se procede a realizar un DataFrame que muestra el precio de las acciones utilizadas en el periodo del 2018-01-31 junto con el porcentaje destinado sobre el capital respecto a la compra de cada activo, asimismo se presenta el total de títulos que se pueden comprar y la comisión que representa cada transacción.

```python
sort_dates = fn.f_sortdates(data_close, dates)

prices = fn.f_prices(data_close, sort_dates)

act_delete = ['KOFL', 'KOFUBL', 'MXN', 'USD', 'NMKA', 'BSMXB']
posdata = fn.f_posdata(act_delete, data_files, files, prices, k, c)
posdata
```

> <font color= 'darkcyan' > <font size = 3> Se hace la obtención del cash o dinero bruto proveniente de los activos que se borraron en la variable llamada "act_delete".

```python
# Cash 
cash = (1-posdata['Peso (%)'].sum())*k
cash
```

> <font color= 'darkcyan' > <font size = 3> Se realiza la sumatoria de las comisiones.

```python
# Summatory of comisions
comision_sum = posdata['Comisiones'].sum()
comision_sum
```

> <font color= 'darkcyan' > <font size = 3> Se obtiene un DataFrame final desde incio hasta el último dato proveniente de nuestra anterior descarga. Posteriormente se hace la distinción entre antes y durante pandemia.

```python
passive_inv = {'Dates': ['2018-01-30'], 'Capital': [k]}
passive_inv = fn.f_passive_inv(sort_dates, posdata, prices, cash, passive_inv)

df_passive = fn.f_passive(passive_inv)
df_passive
```
